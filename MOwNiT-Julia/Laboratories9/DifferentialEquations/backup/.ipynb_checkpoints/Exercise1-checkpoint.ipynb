```julia
# Simple Pendulum Problem
# using OrdinaryDiffEq, Plots
# plotly()

#Constants
const g = 9.81
L = 1.0

#Initial Conditions
# uâ‚€ = [0,Ï€/2]
# poÅ‚oÅ¼enie poczÄ…tkowe, prÄ™dkoÅ›Ä‡ poczÄ…tkowa
uâ‚€ = [7/10*pi,-pi]
# tspan = (0.0,6.3)
tspan = (0.0,80)

#Define the problem
function simplependulum(t,u,du)
    Î¸  = u[1]
    dÎ¸ = u[2]
    du[1] = dÎ¸
    du[2] = -(g/L)*sin(Î¸)
end

#Pass to solvers
prob = ODEProblem(simplependulum,uâ‚€, tspan)
sol = solve(prob,Tsit5())

#Plot
# plot(sol,linewidth=2,title ="Simple Pendulum Problem", xaxis = "Time", yaxis = "Height", label = ["Theta","dTheta"])

```

```julia
function polar2cart(sol;dt=0.02,l1=L,vars=(1))
    u = sol.t[1]:dt:sol.t[end]
    print(u)
    
    p1 = l1*map(x->x[vars[1]], sol.(u))

    x1 = l1*sin.(p1)
    y1 = -l1*cos.(p1)
    (u, (x1 ,
     y1 ))
end
```

```julia
#Obtain coordinates in Cartesian Geometry
ts, ps = polar2cart(sol, l1=L, dt=0.01)
# plot(ps, xlim=(-L*10/7,L*10/7), ylim=(-L,L))
```

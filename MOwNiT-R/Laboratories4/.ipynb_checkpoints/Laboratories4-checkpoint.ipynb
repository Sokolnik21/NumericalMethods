```r
# before You start working on data and creating plots 
# in R, check how it work first:
# http://ggplot2.tidyverse.org/reference/ggplot.html
install.packages("ggplot2")
library("ggplot2")
```

```r
# operating on data that shows comparision evaluation time
# to row size in matrix multiplication functions written in C

# one function is written in naive method and the other
# using BLAS library with applied optimalization
results = read.csv("CalculationTime.csv")
```

```r
# creating vector for naive results and naive mean
naive = results[results$Type == 0,]
avg_naive = aggregate(ClockTime ~ MatrixRows, data=naive, FUN=mean)

# creating vector for blas results and blas mean
blas = results[results$Type == 2,]
avg_blas = aggregate(ClockTime ~ MatrixRows, data=blas, FUN=mean)
```

```r
# adding standard deviation for naive method
avg_naive$sd = aggregate(ClockTime ~ MatrixRows, data=naive, FUN=sd)$ClockTime

# adding standard deviation for blas method
avg_blas$sd = aggregate(ClockTime ~ MatrixRows, data=blas, FUN=sd)$ClockTime
```

```r
# creating pComparision variable to store plot layers
pComparision = ggplot()
```

```r
# adding to pComparision chart with "naive" data which also contains
# error bars

# colour: blue
pComparision = pComparision +
geom_point(data = avg_naive, aes(MatrixRows, ClockTime, colour = "Naive")) +
geom_errorbar(data = avg_naive,
              aes(MatrixRows, ClockTime, ymin = ClockTime - sd, ymax = ClockTime + sd, colour = "Naive"),
              width = 30)
```

```r
# doing the same thing for "blas" data

# colour: red
pComparision = pComparision +
geom_point(data = avg_blas, aes(MatrixRows, ClockTime, colour = "BLAS")) +
geom_errorbar(data = avg_blas,
              aes(MatrixRows, ClockTime, ymin = ClockTime - sd, ymax = ClockTime + sd, colour = "BLAS"),
              width = 30)
```

```r
pComparision
```

```r
# calculating curve fit for naive and blas method
# I suspect that third degree polynomial fits here 
fitBlas = lm(ClockTime ~ poly(MatrixRows, 3, raw=TRUE), data=avg_blas)
fitNaive = lm(ClockTime ~ poly(MatrixRows, 3, raw=TRUE), data=avg_naive)
```

```r
# adding curve fit to blas method
avg_blas$fit = predict(fitBlas, avg_blas)

# adding curve fit to naive method
avg_naive$fit = predict(fitNaive, avg_naive)
```

```r
# adding to pComparision curve fits
pComparision = pComparision +
geom_line(data = avg_blas, aes(MatrixRows, fit, colour = "BLAS")) +
geom_line(data = avg_naive, aes(MatrixRows, fit, colour = "Naive"))
```
